### Sorting the list_antibacterial:
since i keep adding some new compounds, and alternative names, the list needs sorting and check for drop duplicated based on name (but not based on inchikeys since there are so many compounds with alternative names)

```python
list_antibacterial['drug'] = list_antibacterial['drug'].str.lower()
duplicated_drugs = list_antibacterial.duplicated(subset='drug')
list_antibacterial[duplicated_drugs]
```

```python
list_antibacterial = list_antibacterial.drop_duplicates(subset='drug') # no need to drop since there is no duplicated.
len(list_antibacterial)
```

```python
list_antibacterial = list_antibacterial.sort_values(by='drug', ascending=True).reset_index(drop=True)
list_antibacterial.head()
```

```python
len(list_antibacterial)
list_antibacterial.to_csv('list_antibacterial.csv', index=False)
```

### Getting a list of antibacterials to pass into the CC:
* Removes duplicates from list_antibacterial based on InChIKeys
* Keeps only the 'drug' and 'inchikey' columns.
* Converts all drug names to lowercase.
* Sorts the resulting list alphabetically by drug name.

```python
final_list = list_antibacterial.drop_duplicates(subset='inchikey')[['drug', 'inchikey']]
len(final_list)
```

```python
final_list['drug'] = final_list['drug'].str.lower()
final_list = final_list.sort_values(by='drug')
```

### Checking the length of inchikeys:
making sure they all have 27 characters (25 letter)

```python
final_list['inchikey'] = final_list['inchikey'].astype(str).str.strip()
invalid_rows = final_list[final_list['inchikey'].str.len() != 27] 
print(invalid_rows)
```

```python
final_list.to_csv('list_antibacterial_for_cc.csv', index=False)
```
